Q1.What is the difference between a function and a method in Python?

Ans. Difference Between Function and Method in Python

In Python, both functions and methods are used to perform specific tasks, but there’s a clear difference between the two based on where and how they are used.

A function is a block of code that is written once and can be used multiple times anywhere in a program. It is defined using the def keyword and is independent, meaning it doesn’t belong to any object or class. Functions can take inputs (called parameters) and can also return outputs using the return statement.

````python
# Function example
def add(a, b):
    return a + b

result = add(5, 3)
print(result)   # Output: 8
````

Method is also a function but it is associated with an object or a class. In simple words, when a function is defined inside a class, it becomes a method. You call a method using an object of that class. Methods usually work with the data stored inside that particular object.

````python
# Method example
class Student:
    def greet(self):
        return "Hello, I am a student!"

# Creating object
s1 = Student()

# Calling method using object
print(s1.greet())   # Output: Hello, I am a student!
````

Q2.Explain the concept of function arguments and parameters in Python.

Ans.Function Arguments and Parameters in Python

When we define a function, the names listed inside the parentheses of the function definition are called parameters. They act as placeholders that receive values when the function is called.

When we call a function, the actual values we pass into the function are known as arguments. These arguments are assigned to the parameters in order, and they allow the function to perform its operation using those values.

In simple words, parameters are like empty containers that wait to hold some data, while arguments are the actual data that fill those containers when the function runs.

Python allows different types of arguments such as positional arguments, keyword arguments, default arguments, and arbitrary arguments, giving flexibility in how we call and use functions.

<u>Types of Arguments and Parameters in Python</u>

<u>1 Positional Arguments:</u>
These are the most common type. The position of the argument in the function call matters — values are assigned to parameters in the same order they are passed.



````python
def sub(a, b):
    print(a - b)

sub(15, 10)   # Here 15 goes to 'a' and 10 goes to 'b'
````

<u>2.Keyword Arguments:</u>
In this type, we specify which parameter should get which value using the parameter’s name.
The order doesn’t matter because values are matched by name.

````python
def introduce(name, age):
    print("Name:", name, "Age:", age)#here text inside ""gives the name of the parameter

introduce(age=21, name="Nikhil")
````

<u>3.Default Arguments:</u>
These are parameters that already have a default value assigned in the function definition.
If no value is passed during the call, the default value is used automatically.

````python
def greet(name="Guest which is the default fixed value"):
    print("Hello", name)

greet()            # Uses default value "Guest which is the default fixed value"
greet("Nikhil")    # Overrides default value
````

<u>Arbitrary Arguments (*args):</u>
Sometimes we don’t know how many arguments will be passed.
Using *args, we can accept any number of positional arguments in the form of a tuple.

````python
def details(**info):
    print(info)

details(name="Nikhil", age=20, city="Raipur")
````

Q5. What are iterators in Python and how do they differ from iterables?
1. Iterable:

An iterable is any Python object that can be looped over (used in a for loop).
It contains multiple elements but does not give them one by one directly — it needs an iterator to do that.

Common examples of iterables are:
list, tuple, string, set, and dictionary.

Technically, an iterable is any object that implements the __iter__() method, which returns an iterator.

2. Iterator:

An iterator is an object that actually returns elements one by one from an iterable when we call the next() function on it.
It remembers its position, so every time next() is called, it gives the next value until all items are finished.

An iterator is any object that implements both __iter__() and __next__() methods.

````python
# Iterable
numbers = [10, 20, 30]

# Convert iterable to iterator
it = iter(numbers)

print(next(it))  # 10
print(next(it))  # 20
print(next(it))  # 30
````

| **Basis**      | **Iterable**                               | **Iterator**                                    |
| -------------- | ------------------------------------------ | ----------------------------------------------- |
| **Definition** | An object that can return an iterator.     | An object that produces elements one by one.    |
| **Methods**    | Has `__iter__()` method.                   | Has both `__iter__()` and `__next__()` methods. |
| **Example**    | List, Tuple, String                        | Object returned by `iter()` function            |
| **Use**        | Used in loops like `for item in iterable:` | Used to fetch elements manually using `next()`  |


6. Explain the concept of generators in Python and how they are defined.

Generators in Python are a special type of function that allow you to generate values one at a time, instead of storing them all in memory at once. They are mainly used to work with large data efficiently.

A generator is created just like a normal function, but instead of using the return statement, it uses the yield keyword.
Whenever the function with yield is called, it doesn’t execute completely — it pauses and saves its state, returning one value at a time. When called again, it continues from where it left off.

This makes generators memory-efficient because they don’t store all values in memory at once; they produce them on demand.

````python
def simple_generator():
    print("Step 1: Start")
    yield 10

    print("Step 2: Resumed")
    yield 20

    print("Step 3: Resumed again")
    yield 30

# Creating generator object
gen = simple_generator()

print(next(gen))  # Executes till first yield
print(next(gen))  # Continues after first yield
print(next(gen))  # Continues after second yield
````

### **7. What are the advantages of using generators over regular functions?**

Generators in Python provide several advantages compared to regular functions, mainly in terms of efficiency and performance. They are defined using the `yield` keyword and generate values one by one, instead of returning all of them at once like normal functions.

---

1. **Memory Efficiency:**
   Generators don’t store all data in memory. They generate one item at a time, making them ideal for handling large datasets or streams of data.

2. **Lazy Evaluation:**
   Generators produce values only when required, reducing unnecessary computation and saving resources.

3. **Support for Infinite Sequences:**
   They can represent infinite data streams since values are generated on demand, while regular functions would run out of memory.

4. **Better Performance:**
   As values are produced on the fly, generators are faster in iteration and consume less memory compared to functions returning large lists.

5. **Simpler Code:**
   Generators make the code cleaner and easier to manage when working with sequences, without explicitly handling iteration logic.

6. **Automatic State Retention:**
   After each `yield`, the generator remembers its current execution state and resumes from the same point when called again.

---

Generators are especially useful when working with large files, streaming data, or any situation where you don’t need all results at once.

### **10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

All three — `map()`, `reduce()`, and `filter()` — are functional programming tools in Python, but they serve different purposes.

1. **`map()`**

   * Applies a function to **each element** in an iterable.
   * Returns a new iterator with the results.
   * Used for **transformation**.

   ```python
   map(lambda x: x * 2, [1, 2, 3])  # → [2, 4, 6]
   ```

2. **`filter()`**

   * Applies a function that returns `True` or `False` for each element.
   * Only elements that satisfy the condition (i.e., return `True`) are kept.
   * Used for **filtering or selection**.

   ```python
   filter(lambda x: x > 2, [1, 2, 3, 4])  # → [3, 4]
   ```

3. **`reduce()`**

   * Applies a function **cumulatively** to the elements of an iterable, reducing them to a single value.
   * It is imported from the `functools` module.
   * Used for **aggregation**.

   ```python
   from functools import reduce
   reduce(lambda x, y: x + y, [1, 2, 3, 4])  # → 10
   ```

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms

````python
class FibonacciIterator:
    def __init__(self, max_terms):
        if not isinstance(max_terms, int) or max_terms < 0:
            raise ValueError("max_terms must be a non-negative integer.")
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration

        result = self.a

        self.a, self.b = self.b, self.a + self.b

        self.count += 1
        return result

# --- Usage Example ---
manual_iter = FibonacciIterator(4)

print("Manually calling next() for 4 terms:")
print(next(manual_iter))
print(next(manual_iter))
print(next(manual_iter))
print(next(manual_iter))
````

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

````python
def powers_of_two_generator(max_exponent):
    if not isinstance(max_exponent, int) or max_exponent < 0:
        raise ValueError("Exponent must be a non-negative integer.")

    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

# --- Usage ---
print("Powers of 2 up to exponent 6:")
for p in powers_of_two_generator(6):
    print(p)
````

7. Implement a generator function that reads a file line by line and yields each line as a string

````python
def line_reader_generator(filepath):
    with open(filepath, 'r') as file:
        for line in file:
            yield line

FILE_NAME = "temp_data.txt"
with open(FILE_NAME, 'w') as f:
    f.write("Line One\nLine Two\nLine Three")

print("Reading lines:")
for line in line_reader_generator(FILE_NAME):
    print(line.strip())
````
